```python
data=data.replace("'","",regex=True)
data=data.replace(";","",regex=True)
```

```python
data.columns = data.columns.str.replace(' ', '')                 #fixing labels 
data.info()
```

```python
## converting date to datetime format
```

```python
data['Date'] = pd.to_datetime(data['Date'], errors='coerce')
```

```python
data1=data.drop(["Date1","Month"],axis=1)
```

```python
#Other columns in float
for col in data1.columns:
    if data1[col].dtype=="object":
        data1[col]=data1[col].astype(float)
    
```

```python
#This is most challenging task in this dataset. In our raw dataset, Date column is corrupted. 
#We can see only 2022 year throughout the year.But the given data is day wise weather attributes from 2022 to July 2033
#(predicted data). We have to change year from 2022 to 2033.
#First we will assign target_date as '2022-01-01'. Whenever we will find 2022-01-01, we will increment a year. 
#Replace for all rows with a new year,till we encounter '2022-01-01' and continue till the end.
```

```python
target_date = '2022-01-01'                         
target_date = pd.to_datetime(target_date)
index = data1.index[data1['Date'] == target_date].tolist()
year = 2021
max_year = 2033

for i in range(index[0], data1.shape[0]):
    if data1.at[i, 'Date'] == target_date:
        year += 1
        if year > max_year:
            break
    data1.at[i, 'Date'] = data1.at[i, 'Date'].replace(year=year)
```
